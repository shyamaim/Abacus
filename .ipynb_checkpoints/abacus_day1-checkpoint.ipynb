```python
class Abacus:
    def __init__(self):
        self.units = 0
        self.tens = 0

    def set_number(self, num):
        self.units = num % 10
        self.tens = num // 10

    def get_number(self):
        return self.tens * 10 + self.units
    
    def add(self, value):
        print(f"Adding {value} to {self.get_number()}")
        total = self.get_number() + value
        self.set_number(total)
```
