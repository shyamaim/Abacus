```python
class Abacus:
    def __init__(self):
        self.units = 0
        self.tens = 0

    def set_number(self, num):
        self.units = num % 10
        self.tens = num // 10

    def get_number(self):
        return self.tens * 10 + self.units

    def add(self, value):
        print(f"Adding {value} to {self.get_number()}")

        # Step 1: add to units
        self.units += value

        # Step 2: handle carry
        if self.units >= 10:
            carry = self.units // 10
            self.units = self.units % 10
            self.tens += carry

            print(f"Carry occurred: +{carry} to tens")

        print(f"Result → Tens: {self.tens}, Units: {self.units}")
```

```python
a = Abacus()
a.set_number(8)

a.add(7)

print("Final:", a.get_number())
```

```python
class Abacus:
    def __init__(self):
        self.units = 0
        self.tens = 0

    def set_number(self, num):
        self.units = num % 10
        self.tens = num // 10

    def get_number(self):
        return self.tens * 10 + self.units

    def add(self, value):
        print(f"Adding {value} to {self.get_number()}")

        # Add to units
        self.units += value

        # Carry logic
        if self.units >= 10:
            carry = self.units // 10
            self.units = self.units % 10
            self.tens += carry

            print(f"Carry occurred: +{carry} to tens")

        print(f"Result → Tens: {self.tens}, Units: {self.units}")

    def subtract(self, value):
        print(f"Subtracting {value} from {self.get_number()}")

        # Subtract from units
        self.units -= value

        # Borrow logic
        if self.units < 0:
            borrow = (-self.units + 9) // 10
            self.units += borrow * 10
            self.tens -= borrow

            print(f"Borrow occurred: -{borrow} from tens")

        print(f"Result → Tens: {self.tens}, Units: {self.units}")
```

```python
a = Abacus()
a.set_number(12)

a.subtract(7)

print("Final:", a.get_number())
```
